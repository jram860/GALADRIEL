```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
N_BINS = 5000
ENERGY_BOUND = 10 # {MeV}
data_path = "/Users/jram/Geant4/Projects/GA-Filter-Stack/BremsSpec/data/Tungsten_data_02_18_2024/1000um_tungsten.csv"

df = np.genfromtxt(data_path,skip_header=2,delimiter=',')
bin_width = ENERGY_BOUND/N_BINS
print(bin_width)
energy = df[0:,0]
counts = df[0:,1]
print(df[0:,1])


[fig,ax]= plt.subplots()
ax.step(energy[0:1000],counts[0:1000]/bin_width,where='mid')
ax.set_xlabel('energy [MeV]')
ax.set_ylabel(r'dN/dE [MeV$^{-1}$]')
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_title('X-Ray Spectrum [W @ 1000um]')
ax.grid(linestyle = ':',)
plt.show()
```
